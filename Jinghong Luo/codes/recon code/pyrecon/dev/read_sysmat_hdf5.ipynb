```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
top_dir = "../"
fig, axs = plt.subplots(1,3,figsize=(22, 6))
with h5py.File(top_dir+'/data/test_sysmat.hdf5', 'r') as f:
    data = f['sysmat']
    print(data.shape)
    axs[0].imshow(np.sum(data[0],axis=0).T, origin='lower', cmap='gray')
    for img in data:
        axs[1].imshow(img[0].T, origin='lower', cmap='gray',alpha=0.1)
    axs[2].imshow(np.sum(data, axis=(0,1)).T, origin='lower', cmap='gray')
    axtitles=[
        'Fake PPDFs, at zero angle',
        'Fake PPDFs, first detector, stack of all angles, alpha=0.1',
        'Fake PPDFs, sum of all angles'
        ]

    for idx, axtitle in zip(range(0,3),axtitles):
        axs[idx].set_title(axtitle)
    plt.show()
```
